# Izmjene u kodu

## Izvorni Python kod

```python
# Import modula
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Preuzimanje podataka o akcijama kompanije Apple (AAPL)
symbol = 'AAPL'  # Tiker za Apple
data = yf.download(symbol, start='2000-01-01', end='2024-10-31')

# Provera prvih nekoliko redova podataka
print("Prvih pet redova podataka:")
print(data.head())

# Izraunavanje dnevnih povraaja (returns)
data['Daily Return'] = data['Close'].pct_change()

# Izraunavanje kumulativnog povraaja
data['Cumulative Return'] = (1 + data['Daily Return']).cumprod()

# Grafik zatvarajue cene
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Zatvarajua cena')
plt.title('Kretanje cijene akcija Apple-a (2000-2024)')
plt.xlabel('Datum')
plt.ylabel('Cijena (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

## 2) Prikaz kumulativnog povrata

## 3) Prikaz tri grafika - multipliciranje (jednostavnija metoda)

```python
# Import modula
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Izmjena Tikera
symbol = 'TSLA'  # Tiker za Tesla
data = yf.download(symbol, start='2000-01-01', end='2024-10-31')

# Provera prvih nekoliko redova podataka
print("Prvih pet redova podataka:")
print(data.head())

# Izraunavanje dnevnih povrata (returns)
data['Daily Return'] = data['Close'].pct_change()

# Izraunavanje kumulativnog povrata
data['Cumulative Return'] = (1 + data['Daily Return']).cumprod()

# Grafikon 1
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Close'], label='Zatvarajua cijena')
plt.title('Kretanje cijene akcija ')
plt.xlabel('Datum')
plt.ylabel('Vrijednost')
plt.legend()
plt.grid(True)
plt.show()

# Grafikon 2
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Cumulative Return'], label='Kumulativni povrat')
plt.title('Kumulativni povrat investicije (2000-2024)')
plt.xlabel('Datum')
plt.ylabel('Vrijednost')
plt.legend()
plt.grid(True)
plt.show()

# Grafikon 3
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['Daily Return'], label='Dnevni povrat')
plt.title('Dnevni povrat investicije (2000-2024)')
plt.xlabel('Datum')
plt.ylabel('Vrijednost')
plt.legend()
plt.grid(True)
plt.show()
```

## 3) Prikaz tri grafika  - multipliciranje (sloemija metoda)

```python
# Import modula
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Izmjena Tikera
symbol = 'TSLA'  # Tiker za Teslu
data = yf.download(symbol, start='2000-01-01', end='2024-10-31')

# Provera prvih nekoliko redova podataka
print("Prvih pet redova podataka:")
print(data.head())

# Izraunavanje dnevnih povrata (returns)
data['Daily Return'] = data['Close'].pct_change()

# Izraunavanje kumulativnog povrata
data['Cumulative Return'] = (1 + data['Daily Return']).cumprod()

# Prikaz tri grafikona
fig, axs = plt.subplots(3, 1, figsize=(14, 18))

# Zatvarajua cijena
axs[0].plot(data.index, data['Close'], label='Zatvarajua cijena', color='blue')
axs[0].set_title('Kretanje cijene akcija (2000-2024)')
axs[0].set_xlabel('Datum')
axs[0].set_ylabel('Cijena (USD)')
axs[0].legend()
axs[0].grid(True)

# Kumulativni povrat
axs[1].plot(data.index, data['Cumulative Return'], label='Kumulativni povrat', color='green')
axs[1].set_title('Kumulativni povrat investicije (2000-2024)')
axs[1].set_xlabel('Datum')
axs[1].set_ylabel('Kumulativni povrat')
axs[1].legend()
axs[1].grid(True)

# Dnevni povrat
axs[2].plot(data.index, data['Daily Return'], label='Dnevni povrat', color='red')
axs[2].set_title('Dnevni povrat investicije (2000-2024)')
axs[2].set_xlabel('Datum')
axs[2].set_ylabel('Dnevni povrat')
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
